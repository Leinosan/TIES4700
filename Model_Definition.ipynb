```python
# Custom 2-Layer CNN Model
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        # First conv layer: 3 input channels (RGB) -> 32 output channels
        self.conv1 = nn.Conv2d(3, 32, kernel_size=5, padding=2)
        self.pool = nn.MaxPool2d(2, 2)
        # Second conv layer: 32 -> 64 channels
        self.conv2 = nn.Conv2d(32, 64, kernel_size=5, padding=2)

        # After 2 pooling layers: 50 -> 25 -> 12 (rounded down)
        # So final feature map: 64 * 12 * 12 = 9216
        self.fc1 = nn.Linear(64 * 12 * 12, 128)
        self.fc2 = nn.Linear(128, 2)  # Binary classification
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(0.5)

    def forward(self, x):
        # Conv layer 1
        x = self.pool(self.relu(self.conv1(x)))  # 50x50 -> 25x25
        # Conv layer 2
        x = self.pool(self.relu(self.conv2(x)))  # 25x25 -> 12x12
        # Flatten
        x = x.view(-1, 64 * 12 * 12)
        # Fully connected layers
        x = self.dropout(self.relu(self.fc1(x)))
        x = self.fc2(x)
        return x

model = SimpleCNN()
model = model.to(DEVICE)

print(f'\nModel Architecture:')
print(model)
print(f'\nTotal parameters: {sum(p.numel() for p in model.parameters()):,}')

# Loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
```
